```python
# load csv containing viable zipcodes and coordinate pairs
zipcodes_coords_file = ("./resources_mhardy/zipcodes_coordinates.csv")
zipcodes_coords_df = pd.read_csv(zipcodes_coords_file)
zipcodes_coords_df
```

```python
# create list of regionID's for cities we want
good_id_list = zipcodes_coords_df['RegionID'].tolist()
good_id_list
```

```python
time_df
```

```python
# remove rows from time_df that aren't in the list of good locations
for index, row in time_df.iterrows():
    if row['RegionID'] not in good_id_list:
        time_df.drop(index, inplace=True)
        
time_df
```

```python
# recreate transposed dataset
time_transposed = time_df.transpose()
time_transposed.reset_index(drop = False, inplace = True)
time_transposed = time_transposed.rename(columns=time_transposed.iloc[0])
time_transposed = time_transposed.drop([0, 0])
time_transposed = time_transposed.rename(columns={"RegionID": 'DATE'})
time_transposed['DATE'] = pd.to_datetime(time_transposed['DATE'], infer_datetime_format=True)
time_transposed
```

```python
#time_transposed['INFLATION'] = time_transposed.apply(lambda _: '', axis=1)
date_list = time_transposed['DATE'].to_list()
for index, row in cpi_df.iterrows():
    if row['DATE'] not in date_list:
        cpi_df.drop(index, inplace=True)
        
cpi_df= cpi_df.reset_index(drop=True)
cpi_df
```

```python
time_transposed['CPIAUCNS'] = cpi_df['CPIAUCNS']
cols = ['DATE', 'CPIAUCNS'] + [col for col in time_transposed if col != 'CPIAUCNS' and col != 'DATE']
time_transposed = time_transposed[cols]
time_transposed = time_transposed.rename(columns={"CPIAUCNS": "INFLATION"})
time_transposed = time_transposed.dropna()
time_transposed
```
